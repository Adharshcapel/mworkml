```python
#linear regression then ridge and rasso
#house pricing dataset
from sklearn.datasets import load_boston
import numpy as np

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=load_boston()
```

```python
type(df)
```

```python
df
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset
```

```python
dataset['Price']=df.target
```

```python
dataset
```

```python
#dividing the dataset into dependent and independent feature
X=dataset.iloc[:,:-1]# idependent feature from all coloumn we will skip the last coloumn
y=dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
# import nodel
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()

mse=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)

# it takes the model as the parameter and the dataset it is better to
# GIVE TRAIN AND TEST DATA LIKE INPUT X TRAIN AND Y TRAIN
# we apply cross validation in cv we will divide the data in to train and test in such a way that every c
#combination of data is taken by the model and whosever accuracy is better the whole entire thing is combine
```

```python
mse
```

```python
mean_mse=np.mean(mse)
mean_mse
```

```python
#ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV# HELP WITH HYPERPARAMETER TUNNING
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1,1.3,5]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error')
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
```

```python
df=load_breast_cancer()
```

```python
data=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
data['Target']=df.target
data
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model1=LogisticRegression(C=100,max_iter=100)
```

```python
from sklearn.model_selection import GridSearchCV
params={'C':[1,5,10]},{'max_iter':[100,150]}
model2=GridSearchCV(model1,param_grid=params,scoring='f1')
```

```python
model2.fit(X_train,y_train)
```

```python
model2.best_score_
```

```python
model2.best_params_
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
y_pred=model2.predict(X_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
r=r2_score(y_pred,y_test,multioutput='variance_weighted')
print(r)
```
